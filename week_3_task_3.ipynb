```python
from docx import Document
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, PageBreak
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.pagesizes import A4
from reportlab.lib.units import mm
import pandas as pd

items = [
    {"Original":"Summarize this article.",
     "Improved":"Summarize the following article in 5 bullet points, each 12–15 words, and include a one-line takeaway. Article: {{paste article here}}",
     "Observations":"Adds length constraints, output format (bullets + takeaway) and placeholder for the article—makes the task precise."},

    {"Original":"Translate to Spanish.",
     "Improved":"Translate the text to neutral Latin American Spanish. Preserve technical terms and keep formal tone. Text: {{paste text}}",
     "Observations":"Specifies dialect, tone, and how to treat technical terms to avoid ambiguous translations."},

    {"Original":"Write product descriptions.",
     "Improved":"Write three product descriptions (short, medium, long) for a wireless earbud with active noise cancelling. Include features, benefits and a 1-line CTA.",
     "Observations":"Asking for multiple lengths and specific content sections makes outputs directly usable for different channels."},

    {"Original":"Create code to fetch API data.",
     "Improved":"Generate Python code (requests) to GET JSON from 'https://api.example.com/v1/items', handle errors, retry up to 3 times, and show parsed output.",
     "Observations":"Specifies language, endpoint pattern, error handling and retries so the returned code is production-ready."},

    {"Original":"Design a logo.",
     "Improved":"Suggest 5 logo concepts for a vegan cafe named 'Green Spoon'. For each: color palette, shape ideas, and 1-sentence rationale.",
     "Observations":"Converts a vague creative request into structured, actionable concepts for designers."},

    {"Original":"Analyze sales data.",
     "Improved":"Given monthly sales CSV, analyze trends: yoy growth, seasonality, top 5 SKUs, and 3 actionable recommendations. Include sample code for pandas.",
     "Observations":"Gives input format, desired analyses, and asks for code which makes the output directly reproducible."},

    {"Original":"Plan a trip.",
     "Improved":"Create a 7-day itinerary for Kyoto for a budget traveler (₹70,000 total). Include daily plan, transport tips and cost estimates in INR.",
     "Observations":"Adds duration, location, budget, currency and required components for practical planning."},

    {"Original":"Create SQL query.",
     "Improved":"Write an efficient SQL query (PostgreSQL) to find top 10 customers by revenue in 2024, using tables orders(order_id,customer_id,amount,date) and customers(id,name).",
     "Observations":"Specifies SQL dialect, tables and desired result so the assistant can produce precise, runnable SQL."},

    {"Original":"Draft an email.",
     "Improved":"Draft a polite follow-up email (150–200 words) to a hiring manager about interview status, mentioning interview date and expressing continued interest.",
     "Observations":"Sets tone, length, and required content to produce a ready-to-send email."},

    {"Original":"Make a study plan.",
     "Improved":"Create a 6-week study plan for the GATE CSE Discrete Mathematics section: weekly topics, 5 practice problems/day, and 1 weekly mock test.",
     "Observations":"Specifies exam, duration, daily practice load and mock tests—turns vague request into an actionable schedule."}
]


df = pd.DataFrame(items)
display(df)


doc = Document()
doc.add_heading('10 Task-based AI Prompts — Original, Improved, Observations', level=1)
for i, row in enumerate(items, start=1):
    doc.add_heading(f'Prompt {i}', level=2)
    doc.add_paragraph('Original prompt:')
    doc.add_paragraph(row['Original'])
    doc.add_paragraph('Improved prompt:')
    doc.add_paragraph(row['Improved'])
    doc.add_paragraph('Observations:')
    doc.add_paragraph(row['Observations'])
    doc.add_paragraph('')

docx_path = '/content/ai_prompts.docx'
doc.save(docx_path)


pdf_path = '/content/ai_prompts.pdf'
doc_pdf = SimpleDocTemplate(pdf_path, pagesize=A4,
                            rightMargin=20*mm, leftMargin=20*mm, topMargin=20*mm, bottomMargin=20*mm)
styles = getSampleStyleSheet()
title_style = styles['Title']
h2_style = styles['Heading2']
normal = styles['BodyText']

story = []
story.append(Paragraph('10 Task-based AI Prompts — Original, Improved, Observations', title_style))
story.append(Spacer(1, 6))

for i, row in enumerate(items, start=1):
    story.append(Paragraph(f'Prompt {i}', h2_style))
    story.append(Paragraph('<b>Original prompt:</b>', normal))
    story.append(Paragraph(row['Original'], normal))
    story.append(Paragraph('<b>Improved prompt:</b>', normal))
    story.append(Paragraph(row['Improved'], normal))
    story.append(Paragraph('<b>Observations:</b>', normal))
    story.append(Paragraph(row['Observations'], normal))
    story.append(Spacer(1, 8))
    if i % 4 == 0 and i != len(items):
        story.append(PageBreak())

doc_pdf.build(story)

print("Saved:")
print(docx_path)
print(pdf_path)
```
